## Ejercicio 1

Usted es un nutricionista que investiga dos tipos diferentes de dietas para ver si existe una diferencia significativa en la pÃ©rdida de peso despuÃ©s de un mes. Eliges dos grupos aleatorios de personas; un grupo sigue la primera dieta y el otro sigue la segunda. Al final del mes, se registra la pÃ©rdida de peso (en kg) de cada persona:

| Diet 1 | Diet 2 |
|:-------|:-------|
| 2.0 | 3.0 |
| 2.5 | 3.2 |
| 3.0 | 3.1 |
| 2.8 | 2.9 |
| 2.3 | 2.8 |
| 2.7 | 3.0 |
| 2.5 | 3.2 |

Con estos datos se busca responder a la siguiente pregunta: Â¿Existe una diferencia significativa en la pÃ©rdida de peso promedio entre las personas que siguieron la primera dieta y las que siguieron la segunda dieta?

Para sacar conclusiones, siga los siguientes puntos:

-Enunciar la hipÃ³tesis: hipÃ³tesis nula y alternativa.
-Realizar la prueba para comprobar la hipÃ³tesis. Puede utilizar la prueba t de Student.
-Analizar las conclusiones.

```python
import numpy as np
from scipy import stats
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

#HipÃ³tesis nula (Hâ‚€): No hay diferencia en la pÃ©rdida de peso promedio entre las dos dietas. ğ»0:ğœ‡1=ğœ‡2

#HipÃ³tesis alternativa (Hâ‚): Hay una diferencia significativa en la pÃ©rdida de peso promedio entre las dos dietas.
#ğ»1:ğœ‡1â‰ ğœ‡2

diet1 = np.array([2.0, 2.5, 3.0, 2.8, 2.3, 2.7, 2.5])
diet2 = np.array([3.0, 3.2, 3.1, 2.9, 2.8, 3.0, 3.2])

# Prueba t para muestras independientes
t_stat, p_value = stats.ttest_ind(diet1, diet2)

print(f"EstadÃ­stico t: {t_stat}")
print(f"P-valor: {p_value}")
```

Como el p-valor < 0.05, rechazamos la hipÃ³tesis nula.

Hay evidencia estadÃ­stica significativa para afirmar que existe una diferencia en la pÃ©rdida de peso promedio entre las dos dietas despuÃ©s de un mes.


## ANOVA

**ANOVA** (*Analysis of Variance - AnÃ¡lisis de variaciÃ³n*) Es una tÃ©cnica estadÃ­stica utilizada para comparar las medidas de dos o mÃ¡s grupos. La idea detrÃ¡s de ANOVA es descomponer la variabilidad total de los datos en dos componentes: variabilidad entre grupos y variabilidad dentro de los grupos:

- **Variabilidad entre grupos**: Esta variabilidad se refiere a las diferencias entre las medias del grupo. Si esta variabilidad es considerablemente mayor que la variabilidad dentro del grupo, podrÃ­a ser una indicaciÃ³n de que al menos una de las medias del grupo es diferente.
- **Variabilidad dentro del grupo**: Esta variabilidad se refiere a la dispersiÃ³n de los datos dentro de cada grupo. Si todos los grupos tienen una variabilidad similar, entonces cualquier diferencia notable en las medias de los grupos podrÃ­a considerarse significativa.

Las hipÃ³tesis en ANOVA suelen incluir:

- **HipÃ³tesis nula** ($Hâ‚€$): Las medias de todos los grupos son iguales.
- **HipÃ³tesis alternativa** ($Hâ‚$): Al menos una de las medias del grupo es diferente.

Si el resultado de la prueba ANOVA es significativo (por ejemplo, un valor p inferior a un umbral como 0,05), esto sugiere que al menos la media de un grupo es diferente.

## Ejercicio 2

Un agricultor decide probar tres tipos diferentes de fertilizantes para determinar si alguno es superior en tÃ©rminos de producciÃ³n de maÃ­z. El agricultor planta maÃ­z en 15 parcelas idÃ©nticas y utiliza los tres fertilizantes (5 parcelas para cada tipo). Al final de la temporada mide el rendimiento de maÃ­z (en kg) de cada parcela, con el siguiente resultado:

| Fertilizante 1 | Fertilizante 2 | Fertilizante 3 |
|:-------------|:-------------|:-------------|
| 20 | 22 | 24 |
| 21 | 21 | 23 |
| 20 | 23 | 22 |
| 19 | 22 | 23 |
| 20 | 21 | 24 |

Con estos datos busca responder la siguiente pregunta: Â¿Existe una diferencia significativa en el rendimiento promedio del maÃ­z entre los tres tipos de fertilizantes?

Para ayudarte, sigue los siguientes puntos:

-Enunciar la hipÃ³tesis: hipÃ³tesis nula y alternativa.
-Realizar la prueba ANOVA.
-Analizar las conclusiones.
-Si un fertilizante es mejor que otro, Â¿cÃ³mo podemos saberlo?

```python
fert1 = [20, 21, 20, 19, 20]
fert2 = [22, 21, 23, 22, 21]
fert3 = [24, 23, 22, 23, 24]

# Prueba ANOVA de una vÃ­a
f_stat, p_value = stats.f_oneway(fert1, fert2, fert3)

print(f"EstadÃ­stico F: {f_stat}")
print(f"P-valor: {p_value}")
```

ConclusiÃ³n

Como p-valor = 0.00014 < 0.05, rechazamos la hipÃ³tesis nula.
Existe una diferencia significativa en el rendimiento promedio del maÃ­z entre los tres fertilizantes.

```python
# Datos agrupados en una sola lista para Tukey HSD
data = [20, 21, 20, 19, 20, 22, 21, 23, 22, 21, 24, 23, 22, 23, 24]
labels = ['F1'] * 5 + ['F2'] * 5 + ['F3'] * 5

# Crear DataFrame
df = pd.DataFrame({'rendimiento': data, 'fertilizante': labels})

# Prueba de Tukey HSD
tukey = pairwise_tukeyhsd(endog=df['rendimiento'],
                          groups=df['fertilizante'],
                          alpha=0.05)

print(tukey)
```

ConclusiÃ³n

Todos los pares de fertilizantes muestran diferencias estadÃ­sticamente significativas.

El Fertilizante 3 (F3) tiene el mayor rendimiento promedio, seguido por Fertilizante 2, y por Ãºltimo Fertilizante 1.
